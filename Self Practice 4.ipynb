```python
import pandas as pd

# Membaca data harga rumah
df = pd.read_csv('/content/sample_data/Data_Harga_rumah_surabaya.csv', delimiter=';')

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Clean the 'Price' column: remove periods and convert to numeric
df['Price'] = df['Price'].astype(str).str.replace('.', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'])


# Identifikasi data target dan lakukan pelabelan

# Dalam dataset ini, kita akan membuat kolom 'target' dari kolom 'harga'
# dengan membagi harga rumah menjadi 3 kelas numerik (0, 1, 2)

# Misalnya:
# - target 0: Murah
# - target 1: Sedang
# - target 2: Mahal

# Kita bisa menggunakan pd.qcut untuk membagi data menjadi 3 kelompok frekuensi yang sama
df['target'] = pd.qcut(df['Price'], q=3, labels=[0, 1, 2])

# Mapping label numerik ke label teks berdasarkan SOP
label_mapping = {
    0: 'Murah',
    1: 'Sedang',
    2: 'Mahal'
}

# Buat kolom baru 'target_label' yang berisi label teks
df['target_label'] = df['target'].map(label_mapping)

# Tampilkan data dengan label baru
print("\nData dengan label berdasarkan SOP:")
print(df[['Price', 'target', 'target_label']].head())
```
