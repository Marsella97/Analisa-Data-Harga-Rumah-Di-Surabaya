```python
import pandas as pd
import numpy as np


# Muat dataset
df=pd.read_csv('/content/sample_data/Data_Harga_rumah_surabaya.csv', delimiter=';')

# Print column names to verify
print("Column names:", df.columns.tolist())





# Identifikasi jumlah data
print("Jumlah data:", df.shape[0])

# Identifikasi tipe data
print("\nTipe data:")
print(df.dtypes)

# Identifikasi nilai yang hilang
print("\nJumlah nilai yang hilang:")
print(df.isnull().sum())


# Identifikasi outlier (menggunakan IQR)
# Exclude non-numeric columns from outlier detection and outlier correction
numeric_cols = df.select_dtypes(include=np.number).columns
for column in numeric_cols:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  print(f"\nOutlier pada kolom {column}:")
  print(outliers)


# Deskripsi statistik data
print("\nDeskripsi statistik:")
print(df.describe())



# Pembersihan data

# 1. Pengisian nilai yang hilang (jika ada)
# Columns with missing values: Daya Listrik, Ruang Makan, Ruang Tamu, Jumlah Lantai, Hadap, Terjangkau Internet, Lebar Jalan, Sumber Air, Hook, Kondisi Properti
# Example for filling missing values in 'Daya Listrik' with the mean:
# df['Daya Listrik'].fillna(df['Daya Listrik'].mean(), inplace=True)
# Example for filling missing values in a categorical column like 'Ruang Makan' with the mode:
# df['Ruang Makan'].fillna(df['Ruang Makan'].mode()[0], inplace=True)


# 2. Menghapus baris dengan data yang salah (jika ada)
# Example for dropping rows where 'Price' is NaN after conversion:
# df.dropna(subset=[' Price'], inplace=True)


# 3. Mengoreksi nilai outlier
# This is already handled in the outlier identification loop by identifying numeric columns
for column in numeric_cols:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df[column] = df[column].clip(lower_bound, upper_bound)


# Setelah pembersihan data
print("\nDeskripsi statistik setelah pembersihan:")
print(df.describe())
```
